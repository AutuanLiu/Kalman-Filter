```python
uc_map = np.linspace(1e-3, 1e-6, num=10)
```

```python
best_uc = np.zeros(2*2*8)
step = 0
for dtype in ['linear', 'nonlinear']:
    for ndim, vs in zip([5, 10], [true_coefs5, true_coefs10]):
        for i, noise_var in enumerate(noises):
            min_v = float('inf')
            for ucv in uc_map:
                ret = frokf(noise_var, ndim, dtype, eval(f"con_terms_{dtype}{ndim}"), eval(f"con_terms{ndim}"), uc=uc_map[i], trials=2, ntest=10)
                if np.abs(ret - vs).sum() < min_v:
                    min_v = np.abs(ret - vs).sum()
                    print(min_v)
                    best_uc[step] = ucv
            step += 1
```
